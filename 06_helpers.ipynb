## DB Schema

```python
import pandas as pd
from sqlalchemy import create_engine

# Use the exact same DB_URI you used in the previous cell
engine = create_engine('postgresql://postgres:postgres@localhost:5432/shardbox_development')

print("Mapping shardbox-core database schema...\n")

schema_query = """
SELECT 
    table_name, 
    column_name, 
    data_type 
FROM 
    information_schema.columns 
WHERE 
    table_schema = 'public'
ORDER BY 
    table_name, 
    ordinal_position;
"""

df_schema = pd.read_sql(schema_query, engine)

# Display the schema grouped by table for easy reading
for table in df_schema['table_name'].unique():
    print(f"--- Table: {table} ---")
    columns = df_schema[df_schema['table_name'] == table]
    for _, row in columns.iterrows():
        print(f"  - {row['column_name']} ({row['data_type']})")
    print()
```
